# Load Data (features)

```python
import numpy as np
import os
import pickle
```

```python
input_path = '/content/drive/MyDrive/Into the Unknown/dataset/raw_features'

label_slouch1 = []
label_slouch2 = []
label_unbalanced = []

X_arrays = []

lst = os.listdir(input_path)
for people in lst:
  sub_lst = os.listdir(input_path + '/' + people)
  for label in sub_lst:
    path = input_path + '/' + people +'/' + label
    array_lst = os.listdir(path)
    for array_ in array_lst:
      array = np.load(path + '/' + array_)
      X_arrays.append(array)
      if label in ['B', 'B_D']:
        label_slouch1.append(1)
      else:
        label_slouch1.append(0)
      
      if label in ['C', 'C_D']:
        label_slouch2.append(1)
      else:
        label_slouch2.append(0)
      
      if label in ['D', 'B_D', 'C_D']:
        label_unbalanced.append(1)
      else:
        label_unbalanced.append(0)
      

```

```python
X_arrays = np.array(X_arrays)
label_slouch1 = np.array(label_slouch1)
label_slouch2 = np.array(label_slouch2)
label_unbalanced = np.array(label_unbalanced)
```

```python
path = '/content/drive/MyDrive/Into the Unknown/dataset'
save_path = path + '/' + 'X_arrays.npy'
np.save(save_path, X_arrays)

save_path = path +'/' + 'label_slouch1.npy'
np.save(save_path, label_slouch1)

save_path = path +'/' + 'label_slouch2.npy'
np.save(save_path, label_slouch2)

save_path = path +'/' + 'label_unbalanced.npy'
np.save(save_path, label_unbalanced)
```

```python
X_arrays.shape, label_slouch1.shape
```

```python
models_path = '/content/drive/MyDrive/Into the Unknown/models/'
```

# Model for Slouch1

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_arrays, label_slouch1, test_size = 0.2, random_state = 2023)
x_train.shape, y_train.shape
```

```python
x_train
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                   hidden_layer_sizes=(10, 2), random_state=1)

clf.fit(x_train, y_train)

cur_path = models_path + 'slouch1.pkl'
pickle.dump(clf, open(cur_path, 'wb'))
```

```python
clf = pickle.load(open(cur_path, 'rb'))
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, y_pred)
```

```python
clf.score(x_test, y_test)
```
